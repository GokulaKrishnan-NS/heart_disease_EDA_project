```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# this line is used to Set visualization style
sns.set_style("whitegrid")

#this line is used to load the data

print("\n Data Loading and Initial Inspection:")
df = pd.read_csv('heart.csv')

# head function is used to print the first n lines
print("\n First 5 rows:")
print(df.head())

# info function is used to check for data types and null values in the data
print("\n Data Info:")
df.info()

# this line is used to check for null values and number of null values
print("\n Missing values per column:")
print(df.isnull().sum())

# this line is used to tell the statistical analysis and check for unusual values in the dataset

print("\n Descriptive Statistics:")

# Statistics for numerical columns

print("\n Numerical Statistics:")
print(df.describe())

# Exploratory Data Analysis (EDA)

# Target Variable Distribution
plt.figure(figsize=(6, 5))
sns.countplot(x='HeartDisease', data=df, palette='viridis')
plt.title('Distribution of Heart Disease (0=No, 1=Yes)')
plt.xlabel('Heart Disease Status')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['No Disease', 'Disease'])
plt.show()

# Categorical Feature vs. Target (Example: ChestPainType)
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='ChestPainType', hue='HeartDisease', palette='Set1')
plt.title('Heart Disease Status by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No Disease (0)', 'Disease (1)'])
plt.show()

# Numerical Feature Distribution Stacked by Target (Example: Age)
plt.figure(figsize=(9, 6))
sns.histplot(df, x='Age', hue='HeartDisease', multiple='stack', bins=20, kde=True, palette='coolwarm')
plt.title('Age Distribution Stacked by Heart Disease Status')
plt.xlabel('Age (Years)')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['Disease (1)', 'No Disease (0)'])
plt.show()

# Correlation Heatmap (for numerical variables)
plt.figure(figsize=(10, 8))
# this line is used to filter only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=np.number).columns
corr_matrix = df[numerical_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()



print("\n Conclusion:")
print("The analysis revealed strong correlations between HeartDisease and factors such as ChestPainType, MaxHR, and ST_Slope. The visual EDA provides a good foundation for more advanced analysis or model building.")
```
